```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Product_Sell": [10, 15, 18, 22, 26, 30, 5, 31],
    "Revenue_Generation": [1000, 1400, 1800, 2400, 2600, 2800, 700, 2900]
}

df = pd.DataFrame(data)
X = df["Product_Sell"].values
y = df["Revenue_Generation"].values

# Initialize parameters
m = 0  # coefficient (slope)
b = 0  # intercept (bias)
learning_rate = 0.001
epochs = 1000
n = len(X)  # number of data points

# Function to compute the predictions
def predict(X, m, b):
    return m * X + b

# Gradient descent algorithm
for epoch in range(epochs):
    y_pred = predict(X, m, b)
    error = y_pred - y

    # Calculate gradients
    m_gradient = (2/n) * np.dot(error, X)
    b_gradient = (2/n) * np.sum(error)

    # Update parameters
    m -= learning_rate * m_gradient
    b -= learning_rate * b_gradient

    # Print the progress every 100 epochs
    if epoch % 100 == 0:
        loss = np.mean(error ** 2)
        print(f'Epoch {epoch}: m = {m:.4f}, b = {b:.4f}, Loss = {loss:.4f}')

# Final parameters
print(f'Final parameters: m = {m:.4f}, b = {b:.4f}')

# Predictions
y_pred = predict(X, m, b)

# Plot
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Product_Sell')
plt.ylabel('Revenue_Generation')
plt.title('Product_Sell vs Revenue_Generation')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.array([
    [10, 1000],
    [15, 1400],
    [18, 1800],
    [22, 2400],
    [26, 2600],
    [30, 2800],
    [5, 700],
    [31, 2900],
])

# Separate features (predictors) and target variables
X = data[:, 0].reshape(-1, 1)  # Reshape to column vector
y = data[:, 1]

# Initialize weights and learning rate
w = np.random.rand()  # Random weight
b = np.random.rand()  # Random bias
learning_rate = 0.01

# Define the squared loss function
def squared_loss(y_true, y_pred):
  return np.mean(0.5 * (y_true - y_pred) ** 2)

# Perform gradient descent for a certain number of epochs
epochs = 1000
losses = []  # List to store loss values per epoch

for epoch in range(epochs):
  # Predict using current weights and bias
  y_pred = w * X + b

  # Calculate the gradients of the loss function w.r.t. weights and bias
  dw = np.mean(X * (y_pred - y))
  db = np.mean(y_pred - y)

  # Update weights and bias using gradient descent
  w -= learning_rate * dw
  b -= learning_rate * db

  # Calculate and store loss for this epoch
  loss = squared_loss(y, y_pred)
  losses.append(loss)

# Print the final weights and bias (represents the linear regression model)
print(f"Final Weights (w): {w:.4f}")
print(f"Final Bias (b): {b:.4f}")

# Plot the loss vs. epochs
plt.plot(range(epochs), losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss per Epoch (Weight w)")
plt.show()

# Note: This code finds a local minimum, not necessarily the global minimum.
```

```python
import numpy as np
import pandas as pd

# Data
data = {
    "Product_Sell": [10, 15, 18, 22, 26, 30, 5, 31],
    "Revenue_Generation": [1000, 1400, 1800, 2400, 2600, 2800, 700, 2900]
}

df = pd.DataFrame(data)
X = df['Product_Sell'].values
y = df['Revenue_Generation'].values

# Standardize X for better optimization performance
X = (X - np.mean(X)) / np.std(X)

# Add a column of ones to X for the intercept term
X = np.vstack((np.ones(X.shape[0]), X)).T

# Initialize coefficients
w = np.zeros(X.shape[1])

# Hyperparameters
learning_rate = 0.01
iterations = 1000

# Coordinate Descent
for iter in range(iterations):
    for j in range(len(w)):
        # Calculate the partial residual without the current weight
        partial_residual = y - X @ w + X[:, j] * w[j]
        
        # Update the j-th weight
        w[j] = np.sum(partial_residual * X[:, j]) / np.sum(X[:, j] ** 2)

# Final coefficients
intercept, slope = w

# Print the results
print(f"Intercept: {intercept}")
print(f"Slope: {slope}")

# Predicted values
y_pred = intercept + slope * X[:, 1]

# Plotting the results
import matplotlib.pyplot as plt

plt.scatter(df['Product_Sell'], y, color='blue', label='Actual data')
plt.plot(df['Product_Sell'], y_pred, color='red', label='Fitted line')
plt.xlabel('Product_Sell')
plt.ylabel('Revenue_Generation')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Product_Sell": [10, 15, 18, 22, 26, 30, 5, 31],
    "Revenue_Generation": [1000, 1400, 1800, 2400, 2600, 2800, 700, 2900]
}

df = pd.DataFrame(data)
X = df['Product_Sell'].values
y = df['Revenue_Generation'].values

# Standardize X for better optimization performance
X = (X - np.mean(X)) / np.std(X)

# Add a column of ones to X for the intercept term
X = np.vstack((np.ones(X.shape[0]), X)).T

# Initialize coefficients
w = np.zeros(X.shape[1])

# Hyperparameters
iterations = 100

# Lists to store the loss and weights for plotting
losses = []
weights = []

# Coordinate Descent focusing on the slope
for iter in range(iterations):
    # Update slope (weight for Product_Sell)
    partial_residual = y - X @ w + X[:, 1] * w[1]
    w[1] = np.sum(partial_residual * X[:, 1]) / np.sum(X[:, 1] ** 2)
    
    # Compute the loss (mean squared error)
    loss = np.mean((y - X @ w) ** 2)
    losses.append(loss)
    weights.append(w[1])

# Plotting the loss vs. weights
plt.plot(weights, losses, marker='o')
plt.xlabel('Slope (Weight for Product_Sell)')
plt.ylabel('Loss (Mean Squared Error)')
plt.title('Loss vs. Slope during Coordinate Descent')
plt.grid()
plt.show()

# Final coefficients
intercept, slope = w

# Print the results
print(f"Intercept: {intercept}")
print(f"Slope: {slope}")

# Predicted values
y_pred = intercept + slope * X[:, 1]

# Plotting the results
plt.scatter(df['Product_Sell'], y, color='blue', label='Actual data')
plt.plot(df['Product_Sell'], y_pred, color='red', label='Fitted line')
plt.xlabel('Product_Sell')
plt.ylabel('Revenue_Generation')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Product_Sell": [10, 15, 18, 22, 26, 30, 5, 31],
    "Revenue_Generation": [1000, 1400, 1800, 2400, 2600, 2800, 700, 2900]
}

df = pd.DataFrame(data)
X = df['Product_Sell'].values
y = df['Revenue_Generation'].values

# Standardize X for better optimization performance
X = (X - np.mean(X)) / np.std(X)

# Add a column of ones to X for the intercept term
X = np.vstack((np.ones(X.shape[0]), X)).T

# Initialize coefficients
w = np.zeros(X.shape[1])

# Hyperparameters
learning_rate = 0.01
iterations = 100

# Lists to store the loss and weights for plotting
losses = []
weights = []

# Stochastic Gradient Descent (SGD)
for iter in range(iterations):
    # Randomly shuffle the data
    indices = np.random.permutation(X.shape[0])
    X_shuffled = X[indices]
    y_shuffled = y[indices]
    
    # Stochastic Gradient Descent (SGD)
    for i in range(X.shape[0]):
        X_i = X_shuffled[i:i+1]
        y_i = y_shuffled[i:i+1]
        
        # Compute gradient
        gradient = -2 * X_i.T @ (y_i - X_i @ w)
        
        # Update weights
        w -= learning_rate * gradient.flatten()
        
    # Compute the loss (mean squared error)
    loss = np.mean((y - X @ w) ** 2)
    losses.append(loss)
    weights.append(w[1])  # Only store the weight for Product_Sell

# Plotting the loss vs. weights
plt.plot(weights, losses, marker='o')
plt.xlabel('Slope (Weight for Product_Sell)')
plt.ylabel('Loss (Mean Squared Error)')
plt.title('Loss vs. Slope during Stochastic Gradient Descent')
plt.grid()
plt.show()

# Final coefficients
intercept, slope = w

# Print the results
print(f"Intercept: {intercept}")
print(f"Slope: {slope}")

# Predicted values
y_pred = intercept + slope * X[:, 1]

# Plotting the results
plt.scatter(df['Product_Sell'], y, color='blue', label='Actual data')
plt.plot(df['Product_Sell'], y_pred, color='red', label='Fitted line')
plt.xlabel('Product_Sell')
plt.ylabel('Revenue_Generation')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Product_Sell": [10, 15, 18, 22, 26, 30, 5, 31],
    "Revenue_Generation": [1000, 1400, 1800, 2400, 2600, 2800, 700, 2900]
}

df = pd.DataFrame(data)
X = df['Product_Sell'].values
y = df['Revenue_Generation'].values

# Standardize X
X = (X - np.mean(X)) / np.std(X)

# Initialize Parameters
m = 0
b = 0
n = len(X)

# Hyperparameters
learning_rate = 0.01
epochs = 1000
lambda1 = 0.1  # For L1
lambda2 = 0.1  # For L2

# Lists to store the values
m_values = []
b_values = []
loss_values = []

# Function to compute the predictions
def predict(X, m, b):
    return m * X + b

# Gradient Descent with Elastic Net
for epoch in range(epochs):
    y_pred = predict(X, m, b)
    error =  y - y_pred

    # Compute loss with L1 and L2 regularization
    loss = np.mean(error ** 2) + lambda1 * np.sum(np.abs(m)) + lambda2 * np.sum(m ** 2)
    
    # Compute gradients
    m_gradient = -(2/n) * np.dot(error, X) + lambda1 * np.sign(m) + 2 * lambda2 * m
    b_gradient = -(2/n) * np.sum(error)
    
    # Update parameters
    m -= learning_rate * m_gradient
    b -= learning_rate * b_gradient

    # Store the values
    m_values.append(m)
    b_values.append(b)
    loss_values.append(loss)

    if epoch % 100 == 0:
        print(f'Epoch {epoch}: m = {m:.4f}, b = {b:.4f}, Loss = {loss:.4f}')

# Final parameters
print(f'Final parameters: m = {m:.4f}, b = {b:.4f}')

# Predictions
y_pred = predict(X, m, b)

# Plotting the values
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
plt.plot(range(epochs), m_values, label='m (slope)')
plt.xlabel('Epochs')
plt.ylabel('m (slope)')
plt.title('m (slope) over epochs')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(range(epochs), b_values, label='b (bias)', color='green')
plt.xlabel('Epochs')
plt.ylabel('b (bias)')
plt.title('b (bias) over epochs')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(range(epochs), loss_values, label='Loss', color='red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over epochs')
plt.legend()

plt.tight_layout()
plt.show()
```
