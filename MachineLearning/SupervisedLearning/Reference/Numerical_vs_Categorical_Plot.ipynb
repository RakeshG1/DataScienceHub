Bar Chart:

This plot shows the average (or other summary statistic) of the numerical variable for each category of the categorical variable.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {'category': ['A', 'A', 'B', 'B', 'C', 'C'],
        'numerical': [10, 15, 20, 25, 30, 35]}
df = pd.DataFrame(data)

# Create the bar chart (mean by default)
sns.barplot(x="category", y="numerical", data=df)
plt.title("Average Numerical Value by Category")
plt.xlabel("Category")
plt.ylabel("Numerical Value")
plt.show()
```

Box Plot:

This plot shows the distribution of the numerical variable within each category. It includes the median, interquartile range (IQR), and potential outliers.

```python
# Create the box plot
sns.boxplot(x="category", y="numerical", showmeans=True, data=df)
plt.title("Distribution of Numerical Variable by Category")
plt.xlabel("Category")
plt.ylabel("Numerical Value")
plt.show()
```

Violin Plot:

Similar to a box plot, but it shows the density distribution of the data within each category using a violin shape.

```python
# Create the violin plot
sns.violinplot(x="category", y="numerical", data=df) # showmeans=True
plt.title("Distribution of Numerical Variable by Category")
plt.xlabel("Category")
plt.ylabel("Numerical Value")
plt.show()
```

Strip Plot:

This plot shows individual data points along with the median for each category. It's useful for visualizing datasets with many data points where overlapping points in a scatter plot might obscure patterns.

```python
# Create the strip plot
sns.stripplot(x="category", y="numerical", data=df)
plt.title("Individual Values and Median by Category")
plt.xlabel("Category")
plt.ylabel("Numerical Value")
plt.show()
```

Scatter Plot with Hue:

This approach is useful for exploring potential relationships between the numerical variable and individual categories. Color-coding data points by category allows for visual comparison within and across categories.

```python
# Create the scatter plot with hue for category
sns.scatterplot(x="numerical", y="category", hue="category", data=df)
plt.title("Scatter Plot with Category Hue")
plt.xlabel("Numerical Value")
plt.ylabel("Category")
plt.show()
```
