```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
import seaborn as sns

# Generate synthetic data
np.random.seed(11)
X = np.sort((np.random.rand(100, 1) * 10), axis=0)[:62]
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0]) 

X_test = np.sort((np.random.rand(100, 1) * 10), axis=0)[65:74]
y_test = np.sin(X_test).ravel() + np.random.normal(0.45, 0.26, X_test.shape[0])

# Normalize the input features
scaler_X = StandardScaler()
X_normalized = scaler_X.fit_transform(X)
X_test_normalized = scaler_X.transform(X_test)

# Normalize the target variable
scaler_y = StandardScaler()
y_normalized = scaler_y.fit_transform(y.reshape(-1, 1)).ravel()
y_test_normalized = scaler_y.transform(y_test.reshape(-1, 1)).ravel()

# Polynomial transformation function
def polynomial_features(X, degree):
    poly = PolynomialFeatures(degree)
    return poly.fit_transform(X)

# Gradient descent function
def gradient_descent(X, y, degree, learning_rate=0.05, epochs=20000):
    X_poly = polynomial_features(X, degree)
    m = X_poly.shape[1]
    
    # Initialize weights
    W = tf.Variable(tf.random.normal([m, 1], dtype=tf.float64) * 0.01)
    
    # Gradient descent
    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            y_pred = tf.matmul(X_poly, W)
            loss = tf.reduce_mean(tf.square(y - y_pred))

        gradients = tape.gradient(loss, [W])
        W.assign_sub(learning_rate * gradients[0])

        if epoch % 2000 == 0:
            print(f'Epoch {epoch}, Loss: {loss.numpy()}')

    return W

# Train the cubic model using gradient descent
W_cubic = gradient_descent(X_normalized, y_normalized.reshape(-1, 1), degree=3)

# Prediction function
def predict(X, W, degree):
    X_poly = polynomial_features(X, degree)
    return tf.matmul(X_poly, W).numpy()

# Plot predictions
def plot_predictions(X, y, model_weights, degree, line_style, color, label):
    y_pred = predict(X, model_weights, degree)
    y_pred_rescaled = scaler_y.inverse_transform(y_pred.reshape(-1, 1)).ravel()
    plt.plot(X, y_pred_rescaled, linestyle=line_style, color=color, label=label)

# Plot Actual Data and Model Predictions
sns.set_style("dark")
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
plt.scatter(X, y, color="black", label="Train Data")
plt.scatter(X_test, y_test, color="blue", label="Test Data")
plt.xlabel('X-values', fontsize=14)
plt.ylabel('y-values', fontsize=14)
plt.title('Polynomial Regression Models : Predictions', fontsize=16)

# Plot Cubic Model Predictions
plot_predictions(X_normalized, y, W_cubic, degree=3, line_style="-", color="orange", label="Cubic Model")
plot_predictions(X_test_normalized, y_test, W_cubic, degree=3, line_style="dotted", color="orange", label="Cubic Model: Predictions")

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
import seaborn as sns

# Generate synthetic data
np.random.seed(11)
X = np.sort((np.random.rand(100, 1) * 10), axis=0)[:62]
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])

X_test = np.sort((np.random.rand(100, 1) * 10), axis=0)[65:74]
y_test = np.sin(X_test).ravel() + np.random.normal(0.45, 0.26, X_test.shape[0])

# Normalize the input features
scaler_X = StandardScaler()
X_normalized = scaler_X.fit_transform(X)

scaler_X_test = StandardScaler()
X_test_normalized = scaler_X_test.fit_transform(X_test)

# Convert to float32 for compatibility
X_normalized = X_normalized.astype(np.float32)
y = y.astype(np.float32)
X_test_normalized = X_test_normalized.astype(np.float32)
y_test = y_test.astype(np.float32)

# Gradient descent function
def gradient_descent(X, y, degree, learning_rate=0.001, epochs=2000, batch_size=6):
    poly_features = PolynomialFeatures(degree)
    X_poly = poly_features.fit_transform(X)

    # Normalize the polynomial features
    scaler_X_poly = StandardScaler()
    X_poly_normalized = scaler_X_poly.fit_transform(X_poly).astype(np.float32)
    
    # Normalize the output
    scaler_y = StandardScaler()
    y_normalized = scaler_y.fit_transform(y.reshape(-1, 1)).ravel().astype(np.float32)

    # Prepare TF dataset
    train_dataset = tf.data.Dataset.from_tensor_slices((X_poly_normalized, y_normalized))
    train_shuffled_dataset = train_dataset.shuffle(buffer_size=X.shape[0]).batch(batch_size)

    # Initialize weights
    W = tf.Variable(tf.random.normal([X_poly_normalized.shape[1], 1], dtype=tf.float32))

    # Gradient descent
    for epoch in range(epochs):
        for batch_X, batch_y in train_shuffled_dataset:
            with tf.GradientTape() as tape:
                y_pred = tf.matmul(batch_X, W)
                loss = tf.reduce_mean(tf.square(batch_y - tf.squeeze(y_pred)))

            gradients = tape.gradient(loss, [W])
            W.assign_sub(learning_rate * gradients[0])

        if epoch % 100 == 0:
            print(f'Epoch {epoch}, Loss: {loss.numpy()}')
            print(f'Weights: {W.numpy().tolist()}')  # Convert tensor to list and print

    return W, poly_features, scaler_X_poly, scaler_y

# Train models
W_cubic, poly3, scaler_X_poly, scaler_y = gradient_descent(X, y, degree=3, epochs=4000)

# Prediction function
def predict(X, W, poly_features, scaler_X_poly, scaler_y):
    X_poly = poly_features.transform(X)
    X_poly_normalized = scaler_X_poly.transform(X_poly).astype(np.float32)
    y_pred_normalized = tf.matmul(X_poly_normalized, W).numpy()
    y_pred = scaler_y.inverse_transform(y_pred_normalized.ravel().reshape(-1, 1))
    return y_pred

# Plot predictions
plt.scatter(X, y, color='black', label="Train Data")
plt.scatter(X_test, y_test, color='blue', label="Test Data")

y_pred_train = predict(X, W_cubic, poly3, scaler_X_poly, scaler_y)
plt.plot(X, y_pred_train, linestyle="-", color="orange", label="Cubic Model")

y_pred_test = predict(X_test, W_cubic, poly3, scaler_X_poly, scaler_y)
plt.plot(X_test, y_pred_test, linestyle="dotted", color="orange", label="Cubic Model: Predictions")

plt.xlabel('X-values', fontsize=14)
plt.ylabel('y-values', fontsize=14)
plt.title('Polynomial Regression Models : Predictions', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib
from matplotlib import pyplot as plt

matplotlib.rcParams['figure.figsize'] = [9, 6]


x = tf.linspace(-2, 2, 201)
x = tf.cast(x, tf.float32)

def f(x):
  y = x**2 + 2*x - 5
  return y

y = f(x) + tf.random.normal(shape=[201])

plt.plot(x.numpy(), y.numpy(), '.', label='Data')
plt.plot(x, f(x), label='Ground truth')
plt.legend()

class Model(tf.Module):

  def __init__(self):
    # Randomly generate weight and bias terms
    rand_init = tf.random.uniform(shape=[3], minval=0., maxval=5., seed=22)
    # Initialize model parameters
    self.w_q = tf.Variable(rand_init[0])
    self.w_l = tf.Variable(rand_init[1])
    self.b = tf.Variable(rand_init[2])

  @tf.function
  def __call__(self, x):
    # Quadratic Model : quadratic_weight * x^2 + linear_weight * x + bias
    return self.w_q * (x**2) + self.w_l * x + self.b
  
quad_model = Model()

def plot_preds(x, y, f, model, title):
  plt.figure()
  plt.plot(x, y, '.', label='Data')
  plt.plot(x, f(x), label='Ground truth')
  plt.plot(x, model(x), label='Predictions')
  plt.title(title)
  plt.legend()

plot_preds(x, y, f, quad_model, 'Before training')

def mse_loss(y_pred, y):
  return tf.reduce_mean(tf.square(y_pred - y))

batch_size = 32
dataset = tf.data.Dataset.from_tensor_slices((x, y))
dataset = dataset.shuffle(buffer_size=x.shape[0]).batch(batch_size)

# Set training parameters
epochs = 100
learning_rate = 0.01
losses = []

# Format training loop
for epoch in range(epochs):
  for x_batch, y_batch in dataset:
    with tf.GradientTape() as tape:
      batch_loss = mse_loss(quad_model(x_batch), y_batch)
    # Update parameters with respect to the gradient calculations
    grads = tape.gradient(batch_loss, quad_model.variables)
    for g,v in zip(grads, quad_model.variables):
        v.assign_sub(learning_rate*g)
  # Keep track of model loss per epoch
  loss = mse_loss(quad_model(x), y)
  losses.append(loss)
  if epoch % 10 == 0:
    print(f'Mean squared error for step {epoch}: {loss.numpy():0.3f}')

# Plot model results
print("\n")
plt.plot(range(epochs), losses)
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error (MSE)")
plt.title('MSE loss vs training iterations');


plot_preds(x, y, f, quad_model, 'After training')
```

```python
import matplotlib
from matplotlib import pyplot as plt

matplotlib.rcParams['figure.figsize'] = [9, 6]


x = tf.linspace(-2, 2, 150)
x = tf.cast(x, tf.float32)

def f(x):
  y = (1.5 * x**3) + (0.5 * x**2) + (-4 * x) + 0.05
  return y

y = f(x) + tf.random.normal(shape=[150])

plt.plot(x.numpy(), y.numpy(), '.', label='Data')
plt.plot(x, f(x), label='Ground truth')
plt.legend()

class Model(tf.Module):

  def __init__(self):
    # Randomly generate weight and bias terms
    rand_init = tf.random.uniform(shape=[4], minval=-5., maxval=5., seed=22)
    # Initialize model parameters
    self.w_c = tf.Variable(rand_init[0])
    self.w_q = tf.Variable(rand_init[1])
    self.w_l = tf.Variable(rand_init[2])
    self.b = tf.Variable(rand_init[3])

  @tf.function
  def __call__(self, x):
    # Quadratic Model : quadratic_weight * x^2 + linear_weight * x + bias
    return self.w_c * (x**3) + self.w_q * (x**2) + self.w_l * x + self.b
  
quad_model = Model()

def plot_preds(x, y, f, model, title):
  plt.figure()
  plt.plot(x, y, '.', label='Data')
  plt.plot(x, f(x), label='Ground truth')
  plt.plot(x, model(x), label='Predictions')
  plt.title(title)
  plt.legend()

plot_preds(x, y, f, quad_model, 'Before training')

def mse_loss(y_pred, y):
  return tf.reduce_mean(tf.square(y_pred - y))

batch_size = 32
dataset = tf.data.Dataset.from_tensor_slices((x, y))
dataset = dataset.shuffle(buffer_size=x.shape[0]).batch(batch_size)

# Set training parameters
epochs = 100
learning_rate = 0.01
losses = []

# Format training loop
for epoch in range(epochs):
  for x_batch, y_batch in dataset:
    with tf.GradientTape() as tape:
      batch_loss = mse_loss(quad_model(x_batch), y_batch)
    # Update parameters with respect to the gradient calculations
    grads = tape.gradient(batch_loss, quad_model.variables)
    for g,v in zip(grads, quad_model.variables):
        v.assign_sub(learning_rate*g)
  # Keep track of model loss per epoch
  loss = mse_loss(quad_model(x), y)
  losses.append(loss)
  if epoch % 10 == 0:
    print(f'Mean squared error for step {epoch}: {loss.numpy():0.3f}')

# Plot model results
print("\n")
plt.plot(range(epochs), losses)
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error (MSE)")
plt.title('MSE loss vs training iterations');


plot_preds(x, y, f, quad_model, 'After training')
```
