### `What is Gradient?`

- It is the slope (or) rate of change of the function [let's say f(x)] w.r.t x. To find this, we perform differentiation, which is part of the calculus study.
- There are two types of differentiation. i.e.,
  - If x is scalar i.e., x = [1, 2, 4..] Then we perform --> `Single Variable Differentiation, also called as Single Variable Derivative/Gradient`. 
    - In other words, how f(x) rate of change happens w.r.t x.
    - Scalar Derivative Symbol:  [called as f dash / d of / f prime of x]
  - If x is vector i.e., x1 = [1, 2, 4..]; x2 = [4, 3, 6..]; x3 = [4, 12, 43..];...  Then we perform --> `Multi Variable Differentiation, also called as Multiple Variables Derivative/Gradient`.
    - Vector Derivative Symbol:  [called as nabla f of x]
    - In other words, how f(xi) rate of change happens w.r.t x1, x2, x3... individually. This is also called as Partial Differentiation, as each xi affects f(xi) individually, we try to find rate of change or slope of f(xi) individually on each xi. 
    - Partial Derivative Symbol:  [called as partial f by partial x_i]

##### Sample Data

````python
# Generate 30 samples between -2 and 2
x = np.linspace(-2, 1.2, 32)
print("x values:", x)

# Define the cubic polynomial function
y = x**3 - 3*x # y is also called as f(x)
print("y values:", y)
````

#### Find Single Variable: Derivative/Slope/Gradient/Rate of change of f(x) function w.r.t x

- **Using Approximation**
    - This is an approximation method used to estimate the derivative of a function at a specific point. It calculates the slope between two consecutive points on the function.
    - This method is particularly useful when you have discrete data points and want to estimate the rate of change between them.
    - Formula: 

````python
# Calculate finite difference derivatives between all consecutive points
finite_diff_derivatives = np.diff(y) / np.diff(x)

# Plotting the function
plt.figure(figsize=(24, 6))

# Original function
plt.plot(x, y, label='y/f(x) = x^3 - 3x', marker='o')
# Marking the points
# plt.scatter(x, y, color='red', label='Data Points')

# Annotating the points
# for i, (px, py) in enumerate(zip(x, y)):
#     plt.annotate(f"({px:.1f}, {py:.2f})", (px, py), textcoords="offset points", xytext=(0,0), va='top')

# Annotating the slopes
midpoints_x = (x[:-1] + x[1:]) / 2
for i in range(len(midpoints_x)):
    plt.annotate(f"{finite_diff_derivatives[i]:.2f}", 
                (midpoints_x[i], (y[i] + y[i+1])/2), 
                textcoords="offset points", 
                xytext=(0,0), 
                va='top', 
                color='blue',
                size=12)

# Plotting the finite difference lines
for i in range(len(x)-1):
    plt.plot(x[i:i+2], y[i:i+2], color='green', linestyle='--')

plt.xlabel('x values', fontsize=14)
plt.ylabel('y/f(x) values', fontsize=14)
plt.title('Cubic Polynomial Function and Finite Difference Derivatives', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()

# Displaying the x, y, and finite difference derivative values
print("Finite Difference Derivatives between all consecutive points:", finite_diff_derivatives)
````

**Analytical Derivative:**

- This is the exact derivative calculated using calculus. For a given function, ``the analytical derivative gives the precise rate of change at any point on the curve``.
- This method provides the exact rate of change and is derived using differentiation rules.
- As per the derived derivatives formulas/rules, 

````python
# Define the analytical derivative of the cubic polynomial function
y_prime = 3*x**2 - 3

# Midpoints for finite difference derivative plotting
midpoints_x = (x[:-1] + x[1:]) / 2

# Plotting the function and its derivatives
sns.set_style('darkgrid')
plt.figure(figsize=(24, 8))

# Original function
plt.plot(x, y, label='y = x^3 - 3x', marker='o')
# Analytical derivative
plt.plot(x, y_prime, label="Analytical Derivative (y')", linestyle='--', color='red')
# Finite difference derivative
plt.scatter(midpoints_x, finite_diff_derivatives, label="Finite Difference Derivative", color='green')

# Marking the points
plt.scatter(x, y, color='red')

# Annotating the points and the slopes
for i, (px, py) in enumerate(zip(x, y)):
    plt.annotate(f"({px:.1f}, {py:.2f})", (px, py), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)

for i in range(len(midpoints_x)):
    plt.annotate(f"Slope: {finite_diff_derivatives[i]:.2f}", 
                (midpoints_x[i], finite_diff_derivatives[i]), 
                textcoords="offset points", 
                xytext=(0,10 if i % 2 == 0 else -15), 
                ha='center', 
                color='blue', 
                fontsize=8)

plt.xlabel('x values', fontsize=14)
plt.ylabel('y/f(x) values', fontsize=14)
plt.title('Cubic Polynomial Function and Derivatives', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()

# Displaying the values
print("Finite Difference Derivatives between all consecutive points:", finite_diff_derivatives)
print("Analytical Derivative values:", y_prime)
````

**Key Points:**

- By comparing the plots, you will see that the finite difference derivative approximates the analytical derivative but may not be as smooth or accurate, especially in areas where the function changes rapidly.
- Both are differnt are same, where you can see both ways called derivatives values are same. But analytical derivative gives the precise rate of change at any point on the curve on huge data, rather than manual slope calculation between points i.e., derivative approximation, which works better on limited discrete data points.
- Using this slopes between two points of given function, we can find the minima, maxima where the slope of the function is zero and changes direction. It is useful in Gradient Descent algorithms.

### What are Minima, Maxima?

- `These helps showing where the slope of the function is zero and changes direction`.



- Local Minima: In other words, local minima is the place where slope comes to nearly to zero and then suddenly started increasing i.e., on this data point where first derivative is ~ 0 and second derivative i.e., again derivative this point with next point, the slope begins to increase again.   
- Local Maxima: It is the place where slope comes to nearly to zero and then suddenly started decreasing i.e., on this data point where first derivative is ~ 0 and second derivative i.e., again derivative this point with next point, the slope begins to decrease again.   

Intuitively, seeing below plot w.r.t above makes understanding clear.

Mathematicall, calculated local minima and local maxima are:



````python
# Calculate finite difference derivatives between all consecutive points
finite_diff_derivatives = np.diff(y) / np.diff(x)

# Identify local minima and maxima
local_minima = (np.diff(np.sign(np.diff(y))) > 0).nonzero()[0] + 1
local_maxima = (np.diff(np.sign(np.diff(y))) < 0).nonzero()[0] + 1

# Identify global minima and maxima
global_minima = np.argmin(y)
global_maxima = np.argmax(y)

# Plotting the function
plt.figure(figsize=(24, 6))

# Original function
plt.plot(x, y, label='y/f(x) = x^3 - 3x', marker='o')

# Annotating the slopes
midpoints_x = (x[:-1] + x[1:]) / 2
for i in range(len(midpoints_x)):
    plt.annotate(f"{finite_diff_derivatives[i]:.2f}", 
                (midpoints_x[i], (y[i] + y[i+1])/2), 
                textcoords="offset points", 
                xytext=(0,20), 
                va='top', 
                color='blue',
                size=12)

# Plotting the finite difference lines
for i in range(len(x)-1):
    plt.plot(x[i:i+2], y[i:i+2], color='green', linestyle='--')

# Marking local minima and maxima
plt.scatter(x[local_minima], y[local_minima], color='red', label='Local Minima', zorder=5)
plt.scatter(x[local_maxima], y[local_maxima], color='orange', label='Local Maxima', zorder=5)

# Marking global minima and maxima
plt.scatter(x[global_minima], y[global_minima], color='blue', label='Global Minima', zorder=5)
plt.scatter(x[global_maxima], y[global_maxima], color='purple', label='Global Maxima', zorder=5)

# Annotating minima and maxima
for i in local_minima:
    plt.annotate(f"Local Minima\n({x[i]:.2f}, {y[i]:.2f})", (x[i], y[i]), textcoords="offset points", xytext=(0,-20), ha='left', color='red', fontsize=10)

for i in local_maxima:
    plt.annotate(f"Local Maxima\n({x[i]:.2f}, {y[i]:.2f})", (x[i], y[i]), textcoords="offset points", xytext=(0,-50), ha='right', color='orange', fontsize=10)

plt.annotate(f"Global Minima\n({x[global_minima]:.2f}, {y[global_minima]:.2f})", (x[global_minima], y[global_minima]), textcoords="offset points", xytext=(-30,10), ha='left', color='brown', fontsize=10)
plt.annotate(f"Global Maxima\n({x[global_maxima]:.2f}, {y[global_maxima]:.2f})", (x[global_maxima], y[global_maxima]), textcoords="offset points", xytext=(30,-25), ha='right', color='purple', fontsize=10)

plt.xlabel('x values', fontsize=14)
plt.ylabel('y/f(x) values', fontsize=14)
plt.title('Cubic Polynomial Function and Finite Difference Derivatives with Minima and Maxima', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()

# Displaying the x, y, and finite difference derivative values
print("Finite Difference Derivatives between all consecutive points:", finite_diff_derivatives)
````

##### Sample Data

````python
# Sample Data
data = {
    "Product_Sell": [10, 15, 18, 22, 26, 30, 5, 31],
    "Revenue_Generation": [1000, 1400, 1800, 2400, 2600, 2800, 700, 2900]
}

df = pd.DataFrame(data)
X = df['Product_Sell'].values
y = df['Revenue_Generation'].values

df
````

##### `Initial Weights`
```python
m = 0  # coefficient (slope)
b = 0  # intercept (bias)
```

````python
# Standardize X for better optimization performance

# Add a column of ones to X for the intercept term
# X = np.vstack((np.ones(X.shape[0]), X)).T
# print("X --> ", X)

# Initialize Coefficients
# w = np.zeros(X.shape[1]-1) # There is one predictor 
# print("w --> ",w)

# Initialize Parameters
m = 0  # coefficient (slope)
b = 0  # intercept (bias)
n = len(X)  # number of data points

# Function to compute the predictions
def predict(X, m, b):
    return m * X + b

# Hyperparameters Constants
learning_rate = 0.001 # For slower steps size
# iterations = 120
epochs = 100

# Lists to store weights, losses
weights = []
intercepts = []
losses = []
preds = []

print(f'Initial Weights and Intercept: m = {m:.4f}, b = {b:.4f}')

# Gradient Descent (GD) Algorithm
# for i in range(iterations):
for epoch in range(epochs):
    # Compute Prediction Error
    y_pred = predict(X, m, b)
    error = y - y_pred

    # Computer loss with weights and intercept
    loss = np.mean(error ** 2)

    # # Compute partial derivative of loss function w.r.t coefficient. It will find how loss function changes w.r.t existing weight to new weight data point. i.e., slope of the loss function w.r.t weight. 
    # weight_gradient = -2 * X.T @ (y - X @ w) / X.shape[0] # As per above derived derivative formula 
    # # X @ w: Dot Product (Matrix Multiplication) ex: predicted_target = X @ w + b. This tells in general, what is the rate of change in loss function w.r.t this given weight.
    # # Compute partial derivative of loss function w.r.t intercept. It will find how loss function changes w.r.t existing intercept to new intercept data point. i.e., slope of the loss function w.r.t intercept.
    # intercept_gradient = -2 * (y - X @ w) / X.shape[0] # As per above derived derivative formula

    # # Update weights and intercepts
    # # As we wanted to minimize the loss function. i.e., wanted to make near to zero. So with initial weight we would have got some slope. But substracting new slope we move close a little more near to zero. 
    # # Because as slope height reduced due to substract of previous it moves down to zero. How big chunk of slope height/step size decided by learning rate. 
    # # In simple terms, with initial random weights we got some loss function value(A rate of change in loss function w.r.t intial weight). 
    # # And then we continously try to reduce this rate of change in loss function, with intial weights, by substracting its actual weghts and step size step (how big steps) and slope height, 
    # # so that it goes down direction and there is no more rate of change/slope remains in loss function.
    # ## i.e., with latest weights the loss function doesnt have any slope means it is near to zero. In other words, with final weights predictions doesnt make any potential error.
    # weight = weight - learning_rate * weight_gradient # latest weights 
    # intercept = intercept * intercept * weight_gradient # latest intercept

    # Compute partial derivative of loss function w.r.t coefficient & intercept
    # d(MSE) / dw1 = -(2/N) * sum(x1 * (y_true - y_pred))
    m_gradient = (-2/n) * np.dot(error, X) # Dot Product (Matrix Multiplication, then summation) # As per above derived derivative formula 
    b_gradient = (-2/n) * np.sum(error) # As per above derived derivative formula

    # Update latest weights and intercepts
    m = m - learning_rate * m_gradient # latest weights 
    b = b - learning_rate * b_gradient # latest intercept

    # Collect weights, intercept and losses
    weights.append(m)
    intercepts.append(b)
    preds.append(y_pred)
    losses.append(loss)

    # Print the progress every 100 epochs
    # if epoch % 10 == 0:
    print(f'Epoch {epoch}: m = {m:.4f}, b = {b:.4f}, Loss = {loss:.4f}')

# Final parameters
print(f'Final parameters: m = {m:.4f}, b = {b:.4f}')        
````

````python
# Plotting the results 

sns.set_style("dark")
plt.figure(figsize=(18, 5))
# Plotting the values of m, b, and loss over epochs
plt.subplot(1, 3, 1)
plt.plot(range(epochs), weights, label='m (slope)')
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('m (slope)', fontsize=14)
plt.title('m (slope) over epochs', fontsize=14)
plt.legend()
plt.subplot(1, 3, 2)
plt.plot(range(epochs), intercepts, label='b (intercept)', color='green')
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('b (intercept)', fontsize=14)
plt.title('b (intercept) over epochs', fontsize=14)
plt.legend()
plt.subplot(1, 3, 3)
plt.plot(range(epochs), losses, label='Loss', color='red')
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.title('Loss over epochs', fontsize=14)
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
plt.plot(weights, losses, marker='o')
plt.xlabel('Slope (Weight for Product_Sell)', fontsize=14)
plt.ylabel('Loss (Mean Squared Error)', fontsize=14)
plt.title('Loss vs. Weight during Gradient Descent', fontsize=16)
# plt.legend()
# Use FuncFormatter to format y-axis values in non-scientific notation
formatter = FuncFormatter(lambda x, pos: f'{x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)
# Plotting intermediate fitted lines
plt.subplot(1, 3, 2)
plt.scatter(df['Product_Sell'], y, color='blue', label='Actual data')
plt.plot(df['Product_Sell'], preds[0], color='grey', label='Fitted Line: Epoch 1')
plt.plot(df['Product_Sell'], preds[1], color='slategrey', label='Fitted Line: Epoch 2')
plt.plot(df['Product_Sell'], preds[2], color='black', label='Fitted Line: Epoch 3')
plt.xlabel('Product_Sell', fontsize=14)
plt.ylabel('Revenue_Generation', fontsize=14)
plt.title('Actuals vs Fitting Regression Lines [Over Epochs]', fontsize=16)
plt.legend()
# Use FuncFormatter to format y-axis values in non-scientific notation
formatter = FuncFormatter(lambda x, pos: f'{x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)
# Plotting the fitted line
y_pred = predict(X, m, b) # Predictions
plt.subplot(1, 3, 3)
plt.scatter(df['Product_Sell'], y, color='blue', label='Actual data')
plt.plot(df['Product_Sell'], y_pred, color='red', label='Fitted line')
plt.xlabel('Product_Sell', fontsize=14)
plt.ylabel('Revenue_Generation', fontsize=14)
plt.title('Actuals vs Final Regression Line', fontsize=16)
plt.grid()
plt.tight_layout()
plt.show()
````

#### `Key Observations:`

- When we assigned initial weights and intercepts, as m=0 and b=0. The intial loss rate of change/slope value w.r.t this given weight was high and started at left side of optimal w*.
- Then based on smart weight update function i.e., gradient descent(means it try to go in down direction irrespective of left/right side of the hill in other words intial random weights), then it slowly goes down by reducing slope height by minus its intial slope height * step size(meaning how much big jump we wanted to go down). Like this its go down to optimal w*. Where with w* the slope height/rate of change in loss function is near to zero. So prediction error doesnt have any huge potential errors.
- Note: This is the simple curve is related to weights updation function, to see how each time weight update changes in loss function value. And helps to find optimal w*. Just like elbow curve in clustering. But here we go down direction by using its intial rate of change/slope height and slowing reducing it.
- This is how, Gradient Descent works and helps to find optimal w* and intercept in regression model to make least loss value.
