### Sample Data

### What is log/logarithm?

- A logarithm is a mathematical function, it answers the question "To what power must the base be raised to obtain the given number?"

```
logb(x)=y
```
This equation can be interpreted as: b raised to the power of y equals x.

For example,

- log10(100) = 2, because 10^2 = 10*10 = 100, here 10 is the base, 100 is given number 2 is the exponent to which 10 must be raised to produce 100.
- log2(8) = 3, because 2^3 = 2*2*2 = 8, here 2 is the base, 8 is given number 3 is the exponent to which 10 must be raised to produce 8.

We use log2 when working with binary-based quantities or measurements, and log10 when working with quantities or measurements that naturally follow powers of 10.

Log2 Usage:

- ``Exponential Growth and Decay``: As Log2 tells 2 to the power of desired exponent value, tells how much power that value as increased i.e., exponential/exponent growth.

- ``Handle Skewness``: Logarithmic transformation is a powerful technique used in data analysis to address issues like skewness, particularly in datasets exhibiting exponential growth or decay.

### Exponential Data
    - Exponential data refers to a type of data distribution where the values increase or decrease rapidly as a function of another variable. In an exponential distribution, the rate at which the values change is proportional to the current value. This results in a curve that starts slowly and then accelerates or decelerates as it progresses.
    - A common example of exponential data is population growth or decay. In population growth, the number of individuals increases exponentially over time, especially when resources are abundant.
- ``Skewed Data Distribution``:
    - Exponential growth data often results in a skewed distribution, where a few extreme values dominate the dataset, making it challenging to interpret or analyze effectively. Skewness can affect the accuracy of statistical models and visualizations, as well as violate assumptions of normality required for certain analyses.
- ``Logarithmic Transformation``:
    - Logarithmic transformation involves taking the logarithm of each data point in the dataset. In the case of exponential growth data, applying a logarithmic transformation can compress the scale of the data, ``reducing the influence of extreme values and making the distribution more symmetric. This transformation helps spread out the extreme values and brings them closer to the rest of the data, resulting in a more balanced distribution.``.
- ``Mitigating Skewness``:
    - By transforming the exponential growth data using logarithms, skewness is reduced, and the data distribution becomes more symmetric and easier to analyze. Log-transformed data often adheres better to the assumptions of statistical tests and models, enabling more accurate inference and interpretation.
    - Additionally, log transformation can stabilize the variance of the data, making it more homoscedastic, which is beneficial for regression analysis and other modeling techniques.

#### Mathematical Notation to original Integar value

````python
# Sample exponential data
np.random.seed(0)
sample_data = np.random.exponential(scale=1, size=200)

# Plot Exponential Data
sns.set_style("dark")
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Exponential Data Distribution
sns.histplot(sample_data, bins=30, color='steelblue', alpha=0.7, kde=True, ax=axes[0])
axes[0].set_title("Exponential Data => Skewed Data Distribution")
axes[0].set_xlabel("Values")
axes[0].set_ylabel("Density")

# Log Transformed Data = Normal/Symmetric Data Distribution
transformed_sample_data = np.log2(sample_data) #math.log2(sample_data)
sns.histplot(transformed_sample_data, bins=30, color='teal', alpha=0.7, kde=True, ax=axes[1])
axes[1].set_title("Log Transformed Data => Normal/Symmetric Data Distribution")
axes[1].set_xlabel("Values")
axes[1].set_ylabel("Density")

plt.tight_layout()
plt.show()
````
