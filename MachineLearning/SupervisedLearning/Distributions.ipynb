### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**``Normal Distribution``**: 
- Bell-shaped and symmetric, centered around the mean, with most values falling close to the mean. ``Used for data that clusters around a central value``.

**``Exponential Distribution``**: 
- Right-skewed, models time until an event occurs, ``probability decreases exponentially``. Used for modeling time between events.

**``Uniform Distribution``**: 
- ``Flat and constant probability, all values within the range are equally likely``. Used for generating random numbers with no preference for any subset of the range.

```python
samples_normal = np.random.normal(loc=0, scale=1, size=1000)
samples_exponential = np.random.exponential(scale=1, size=1000)
samples_uniform = np.random.uniform(low=0, high=1, size=1000)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(samples_normal, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution')

plt.subplot(1, 3, 2)
plt.hist(samples_exponential, bins=30, density=True, alpha=0.6, color='b')
plt.title('Exponential Distribution')

plt.subplot(1, 3, 3)
plt.hist(samples_uniform, bins=30, density=True, alpha=0.6, color='r')
plt.title('Uniform Distribution')

plt.tight_layout()
plt.show()
```
