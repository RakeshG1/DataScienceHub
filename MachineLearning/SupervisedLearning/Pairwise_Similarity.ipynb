### Import Libraries

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Sample Data

```python
# Create a sample dataset
data = np.array([
    [1, 2],
    [3, 4],
    [5, 6],
    [7, 8]
])

data
```

### Pair Wise Distance Calculation 

```python
# Compute pairwise distances using Euclidean metric
pairwise_distances = pdist(data, metric='euclidean')
print(pairwise_distances)

# Convert to a square matrix
distance_matrix = squareform(pairwise_distances)

print("Pairwise Distance Matrix:\n", distance_matrix)
```

### Visualise

```python
# Convert the distance matrix to a DataFrame for better visualization
distance_df = pd.DataFrame(distance_matrix, 
                            columns=['Point1', 'Point2', 'Point3', 'Point4'], 
                            index=['Point1', 'Point2', 'Point3', 'Point4'])

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(distance_df, annot=True, cmap='viridis')
plt.title('Pairwise Distance Matrix')
plt.show()
```

Observation

- Here, we can observe that 1st data point is near to 2nd data point, and bit far to 3rd data point, and very far from 4th data point, which is correct.
