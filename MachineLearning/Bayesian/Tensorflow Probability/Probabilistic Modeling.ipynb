### Prepare Normal Distribution

```python
# Define a Normal Distribution
normal_dist = tfp.distributions.Normal(loc=0.0, scale=1.0)
```

```python
normal_dist
```

### Get Sample from Normal Distribution

```python
# Sample from the distribution
samples = normal_dist.sample(5)
print("Samples:", samples.numpy())
```

### Compute Probability Density for a Value

```python
# Compute the probability density for a value
prob = normal_dist.prob(0.0)
print("Probability Density at 0:", prob.numpy())
```

```python
# Compute the probability density for a value
prob = normal_dist.prob(2.0)
print("Probability Density at 2:", prob.numpy())
```

**Key Point**: The probability density function (PDF) does not give the probability directly. Instead:

* The height of the curve at a point represents how dense the probability is near that value.
* A higher density near 0 means values around 0 are more likely.
* A lower density at 2 means values around 2 are less likely.
