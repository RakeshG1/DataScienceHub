### Linearity vs. Non-Linearity in Time Series

**Simple Explanation:**

- **Linearity**: In a linear time series, the relationship between the time points is a straight line or can be represented by a linear equation. The changes in the series follow a consistent pattern over time. Linear models assume that the future values are a linear combination of past values.
  
- **Non-Linearity**: In a non-linear time series, the relationship between the time points is not a straight line. The changes in the series do not follow a consistent pattern and can be more complex. Non-linear models can capture more complex relationships and patterns that linear models cannot.

### Summary

- **Linear Time Series**: Captures straight-line trends and consistent patterns over time. Example: a linear trend with noise.
- **Non-Linear Time Series**: Captures more complex, inconsistent patterns over time. Example: a sine wave pattern with noise.

These examples illustrate the difference between linear and non-linear time series. In practice, identifying whether a time series is linear or non-linear helps in selecting appropriate models for analysis and forecasting.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a linear time series with some noise
np.random.seed(0)
time = np.arange(100)
linear_trend = 2 * time + np.random.normal(size=100)

# Generate a non-linear time series (sine wave with noise)
nonlinear_trend = np.sin(0.1 * time) + np.random.normal(scale=0.1, size=100)

# Plot the linear time series
# plt.figure(figsize=(10, 5))
sns.set_style("dark")
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].plot(time, linear_trend, label='Linear Trend with Noise')
axes[0].set_title('Linear Time Series')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Value')
# Plot the non-linear time series
axes[1].plot(time, nonlinear_trend, label='Sine Wave with Noise')
axes[1].set_title('Non-Linear Time Series')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Value')
# plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
