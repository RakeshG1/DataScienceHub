### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial import distance

%matplotlib inline
```

<!--  -->

### Single Vector

```python
# Define Vector
vector_x, vector_y = 3, 4

# Plot
sns.set_style("dark")
plt.figure(figsize=(5,4))
arrow = plt.arrow(0, 0, vector_x, vector_y, 
                    head_width=0.2, head_length=0.3, color="black")
plt.xlabel("X-axis", size=10)
plt.ylabel("Y-axis", size=10)
plt.title("Single Vector", size=14)
plt.grid(True)
plt.show()
```

<!--  -->

### Sum of Vectors

```python
# Define Vectors
vect_1 = (5, 4)
vect_2 = (2, 1)

# Sum of Vectors
sum_vectors = tuple(map(lambda x, y: x + y, vect_1, vect_2))
print("sum_vectors %s" % str(sum_vectors))

# Plot
sns.set_style("dark")
plt.figure(figsize=(5,4))
plt.arrow(0, 0, vect_1[0], vect_1[1], head_width=0.2, head_length=0.3, color="black", label="vector1")
plt.arrow(0, 0, vect_2[0], vect_2[1], head_width=0.2, head_length=0.3, color="black", label="vector2")
plt.arrow(0, 0, sum_vectors[0], sum_vectors[1], head_width=0.2, head_length=0.3, color="blue", label="summed vector")
plt.xlabel("X-axis", size=10)
plt.ylabel("Y-axis", size=10)
plt.title("Single Vector", size=14)
plt.grid(True)
plt.show()
```

Observation

- The sum combines the directions and magnitude of the two original vectors into a single new vector.

### Dot Product of Vectors

```python
# Define vectors
vect_1 = (5, 4)
vect_2 = (2, 1)

# Dot product
dot_product = sum(x * y for x, y in zip(vect_1, vect_2))
print("Dot Product:", dot_product)
```

```python
# Define vectors
vect_1 = (5, 4)
vect_2 = (3, 3)

# Dot product
dot_product = sum(x * y for x, y in zip(vect_1, vect_2))
print("Dot Product:", dot_product)
```

```python
# Define vectors
vect_1 = (5, 4)
vect_2 = (5, 3)

# Dot product
dot_product = sum(x * y for x, y in zip(vect_1, vect_2))
print("Dot Product:", dot_product)
```

Observation

- Higher dot product value, means two vectors are in more same direction

### Magnitude of Vector
- It is euclidean distance between origin & vector

```python
def get_magnitude(vector):
    """This functions calculates vector magnitude"""
    x, y = vector
    return round(np.sqrt(x**2 + y**2),2)
```

```python
# Define vector
vect_1 = (5, 4)

# Get vector magnitude
vect_magnitude = get_magnitude(vect_1)
print("Vector Magnitude --> {0}".format(vect_magnitude))

# Euclidean distance
dst = round(distance.euclidean((0, 0), vect_1), 2)
print("Euclidean distance --> {0}".format(dst))

# Plot
sns.set_style("dark")
plt.figure(figsize=(5,4))
arrow = plt.arrow(0, 0, vect_1[0], vect_1[1], 
                    head_width=0.2, head_length=0.3, color="black")
plt.xlabel("X-axis", size=10)
plt.ylabel("Y-axis", size=10)
plt.title("Magnitude of Vector", size=14)
plt.grid(True)
plt.show()
```
