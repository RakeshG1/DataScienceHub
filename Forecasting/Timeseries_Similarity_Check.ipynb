```python
no_of_records = 24

date_value = np.array("2022-01-01", dtype=np.datetime64)
# Creating 60 Dates, Starting From 2022-01-01
date_values = date_value + np.arange(no_of_records)
date_values = date_values.astype('str')
print("--"*10)
print("Dates")
print("--"*10)
display(date_values)
```

```python
# Sample Data 
data_1 = pd.Series([1,3,6,8,5,12,18,6,5,3,6,1,10,3,6,2,7,12,20,6,5,3,12,1])
data_2 = pd.Series([1,3,6,4,3,12,14,6,5,3,2,0,5,3,2,2,5,12,18,6,5,3,6,1])


print("Length of data 1 : ", len(data_1))
print(data_1.tolist())
print("Length of data 2 : ", len(data_2))
print(data_2.tolist())
```
